```python
"""
=================
Stacked Bar Graph
=================

Example for creating a stacked bar plot.
"""


import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd



onco = pd.read_csv('PhenotypeStacked.csv')

plt.style.use('seaborn')

oncogenes = onco.iloc[:,1:6]
PURO = onco.PURO
MYC = onco.MYC
HRAS = onco.HRAS
CYCD = onco.CYCD
AURKB = onco.AURKB


plt.style.use('seaborn')

N = 5
bp1 = (oncogenes.iloc[0,:])
bp2 =(oncogenes.iloc[1,:])
bp3 = (oncogenes.iloc[2,:])
bp4 = (oncogenes.iloc[3,:])
bp5 = (oncogenes.iloc[4,:])
bp6 = (oncogenes.iloc[5,:])

ind = np.arange(N)    # the x locations for the groups
width = 0.5       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, bp1, width, color = 'greenyellow')
p2 = plt.bar(ind, bp2, width,
             bottom=bp1, color = 'limegreen')
p3 = plt.bar(ind, bp3, width,
             bottom = bp1+bp2, color = 'green')
p4 = plt.bar(ind, bp4, width,
             bottom = bp1+bp2+bp3, color = 'lightskyblue')
p5 = plt.bar(ind, bp5, width, 
             bottom = bp1+bp2+bp3+bp4, color = 'dodgerblue')
p6 = plt.bar(ind, bp6, width, 
             bottom = bp1+bp2+bp3+bp4+bp5, color = 'blue')

plt.ylabel('% of cells', size = 40)
#plt.title('MCF10A oncogene spindle \n phenotype distributions', size = 40)
plt.xticks(ind, ('PURO', 'HRAS', 'MYC', 'CYCD', 'AURKB'), size = 40)
plt.yticks(np.arange(0, 110, 10), size = 35)
#plt.legend(onco.iloc[:,0],
#         bbox_to_anchor=(1., .100),loc=2, borderaxespad=0. 
#         #, prop={'size': 30})
plt.show()




















```
